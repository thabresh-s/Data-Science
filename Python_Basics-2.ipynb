Question 8 - Write a program to calculate cost price if selling price and profit percentage are given:<br>
`CP = ( SP * 100 ) / ( 100 + percentage profit)`





```
sp=float(input("enter the selling price "))
pp=float(input("enter the profit percentage "))
cp=(sp*100)/(100+pp)
print("The cost price would be ",cp)
```

Question 9 - Write a program to calculate cost price if selling price and loss percentage are given:<br>
`CP = ( SP * 100 ) / ( 100  percentage loss )`





```
sp=float(input("enter the selling price "))
lp=float(input("enter the loss percentage "))
cp=(sp*100)/(100-lp)
print("The cost price would be ",cp)
```
