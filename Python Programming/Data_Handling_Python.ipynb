Question 4. Write the program to calculate the amount payable if money has been lent on simple interest.

Hint :- 
  1. Simple Interest(SI) = (P*R*T)/100
  2. Amount Payable = P + SI

P,R,T are given as input to the program


```
p = float(input("Enter the principal amount: "))
r = float(input("Enter the rate of interest: "))
t = float(input("Enter the time: "))
si = (p*r*t)/100
print("The simple interest is ",si)
np=p+si
print("The Net Payable amount is ",np)
```
